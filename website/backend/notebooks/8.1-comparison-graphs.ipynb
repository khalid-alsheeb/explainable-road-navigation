```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v2 = np.array([11, 11])
anytime = np.array([15, 5])
xpoints = np.array([0, 1000])
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(xpoints, v2, label = 'DP (V2) Algorithm')
ax.plot(xpoints, anytime, label = 'Anytime Algorithm')

ax.grid(True)
ax.set_xlabel('Iterations', size=15)
ax.set_ylabel('ISP value', size=15)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(loc=1, prop={'size': 16})
plt.title('ISP optimal value for each algorithm', size=17)

plt.show()
```
